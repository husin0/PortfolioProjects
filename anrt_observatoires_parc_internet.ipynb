```python
import pandas as pd
import csv
import sqlite3
import matplotlib.pyplot as plt
import numpy as np
```

```python
#cleaning Data
df = pd.read_excel('Documents/anrt_observatoires_parc_internet.xlsx')
df = pd.read_excel('Documents/anrt_observatoires_parc_internet.xlsx', skiprows=lambda x: 0<=x<=18)
df.replace(np.nan, 0, inplace=True)
```

```python
#ADSL network in Morocco

zero = 0
num = 0


ADSL_maroc_telecom_2 = []
while zero < 18:
    ADSL_maroc_telecom = df["A"]
    ADSL_maroc_telecom_1 = ADSL_maroc_telecom[num]
    ADSL_maroc_telecom_2.append(ADSL_maroc_telecom_1)
    zero += 1
    num += 4
    
zero = 0
num = 0    
ADSL_medi_telecom_orange_2 = []
while zero < 18:
    ADSL_medi_telecom_orange = df["B"]
    ADSL_medi_telecom_orange_1 = ADSL_medi_telecom_orange[num]
    ADSL_medi_telecom_orange_2.append(ADSL_medi_telecom_orange_1)
    zero += 1
    num += 4
    
zero = 0
num = 0    
ADSL_wana_2 = []
while zero < 18:
    ADSL_wana = df["C"]
    ADSL_wana_1 = ADSL_wana[num]
    ADSL_wana_2.append(ADSL_wana_1)
    zero += 1
    num += 4
    
zero = 0
num = 0    
ADSL_global_2 = []
while zero < 18:
    ADSL_global = df["D"]
    ADSL_global_1 = ADSL_global[num]
    ADSL_global_2.append(ADSL_global_1)
    zero += 1
    num += 4

years =range(2006, 2024) 



data_frame = pd.DataFrame({'ADSL_maroc_telecom_2': ADSL_maroc_telecom_2,
                   'ADSL_medi_telecom_orange_2': ADSL_medi_telecom_orange_2,
                  'ADSL_wana_2': ADSL_wana_2,
                  'ADSL_global_2': ADSL_global_2}, index=years)
ax = data_frame.plot.bar()
```
